```python
test_input = """
    [D]    
[N] [C]    
[Z] [M] [P]
 1   2   3 

move 1 from 2 to 1
move 3 from 1 to 3
move 2 from 2 to 1
move 1 from 1 to 2
"""
```

```python
from dataclasses import dataclass

@dataclass
class Move:
    count: int = 0
    src: int = 0
    dst: int = 0

def parse_p1(x):
    stack, commands = x.split("\n\n")

    def parse_cmds(cmds):
        out = []
        lns = cmds.split("\n")
        for ln in lns:
            _, c, _, s, _, d = ln.split(" ")
            out.append(Move(count=int(c), src=int(s), dst=int(d)))
        return out
    

    def parse_stack(stack):
        # num stacks
        lns = stack.strip().split("\n")
        nums = lns[-1].strip().split("   ")
        nstacks = max(map(int, nums))
        print("Num stacks:", nstacks)

        crates = []
        lns = stack.split("\n")
        for ln in lns[:-1]:
            rowsz = len(ln)
            if rowsz == 0:
                continue
            print(nstacks, rowsz)
            assert (nstacks * 3) + (nstacks - 1) == rowsz

            line = []
            for idx in range(nstacks):
                lo, hi = idx * 4, idx * 4 + 4
                chrs = ln[lo: hi]

                if chrs.strip() == "":
                    line.append(None)
                else:
                    char = chrs.strip().strip("[]")
                    line.append(char)

            crates.append(line)

        return crates

    stack = parse_stack(stack)

    cmds = parse_cmds(commands.strip())
    return stack, cmds

parse_p1(test_input)
```

```python
from collections import deque
def build_deque(stack):
    out = {}
    height, width = len(stack), len(stack[0])
    for idx in range(width):
        # walk down columns
        tmp: deque = deque()
        for jdx in range(height):
            if stack[jdx][idx] is None:
                continue
            tmp.append(stack[jdx][idx])

        out[idx+1] = tmp

    return out
stack, cmds = parse_p1(test_input)
```

```python
def solve_p1(stack, commands):
    q = build_deque(stack)
    print(q)
    for cmd in commands:
        cmd: Move
        sq: deque = q[cmd.src]
        dq: deque = q[cmd.dst]

        for _ in range(cmd.count):
            el = sq.popleft()
            dq.appendleft(el)
            

        print(q)
    
    ss = sorted(q.items(), key=lambda x: x[0])
    ans = ''
    for idx, q in ss:
        ans += q.popleft()

    return ans

solve_p1(*parse_p1(test_input))
```

```python
def solve_p2(stack, commands):
    q = build_deque(stack)
    print("INIT")
    print(q)
    for cmd in commands:
        cmd: Move
        print(cmd)
        sq: deque = q[cmd.src]
        dq: deque = q[cmd.dst]

        if cmd.src == cmd.dst:
            raise Exception("")

        if cmd.count == 0:
            raise Exception("")

        tq = deque()
        for _ in range(cmd.count):
            el = sq.popleft()
            tq.appendleft(el)

        for _ in range(cmd.count):
            el = tq.pop()
            sq.appendleft(el)

        for _ in range(cmd.count):
            el = sq.popleft()
            dq.appendleft(el)
            
        print("RESULT")
        print(q)
    
    ss = sorted(q.items(), key=lambda x: x[0])
    ans = ''
    for idx, q in ss:
        ans += q.popleft()

    return ans

solve_p2(*parse_p1(test_input))
```
