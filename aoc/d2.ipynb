```python
test_input = """
A Y
B X
C Z
"""

def parse_p1(t):
    lns = t.strip().split("\n")
    return [ln.strip().split(" ") for ln in lns]

parse_p1(test_input)
```

```python
from enum import Enum, auto


translate = {
    'X': 'A',
    'Y': 'B',
    'Z': 'C'
}

class Outcomes(Enum):
    WIN = 6
    LOSE = 0
    DRAW = 3

values = {
    'A': 1,
    'B': 2,
    'C': 3
}
def determine_outcome(x, y):
    y = translate[y]

    if x == y:
        return Outcomes.DRAW
    
    if x == 'A' and y == 'B':
        return Outcomes.WIN
    elif x == 'B' and y == 'A':
        return Outcomes.LOSE
    elif x == 'A' and y == 'C':
        return Outcomes.LOSE
    elif x == 'C' and y == 'A':
        return Outcomes.WIN
    elif x == 'C' and y == 'B':
        return Outcomes.LOSE
    elif x == 'B' and y == 'C':
        return Outcomes.WIN
    else:
        raise Exception(f"{x}, {y}")

def solve_p1(lns):
    total = 0
    for ln in lns:
        outcome = determine_outcome(*ln)
        sel = translate[ln[-1]]
        val = values[sel] + outcome.value
        total += val

    return total

solve_p1(parse_p1(test_input))
```

```python
tmp = {
    'X': Outcomes.LOSE,
    'Y': Outcomes.DRAW,
    'Z': Outcomes.WIN
}

beats = {
    'A': 'B',
    'B': 'C',
    'C': 'A'
}
loses = { v: k for k,v in beats.items() }
def determine_choice(p1, outcome):
    if outcome == Outcomes.DRAW:
        return p1
    
    elif outcome == Outcomes.WIN:
        return beats[p1]
    else:
        return loses[p1]

def solve_p2(lns):
    total = 0
    for ln in lns:
        outcome = tmp[ln[-1]]
        choice = determine_choice(ln[0], outcome)
        val = values[choice] + outcome.value

        print(ln, outcome, choice, val)
        total += val
    return total

solve_p2(parse_p1(test_input))
```
