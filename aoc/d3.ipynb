```python
def parse_p1(x):
    lns = x.strip().split("\n")
    parts = []
    for ln in lns:
        s = len(ln)
        left, right = ln[:s//2], ln[s//2:]
        assert s % 2 == 0
        parts.append(
            (list(left), list(right))
        )
    return parts


parse_p1(test_input)
```

```python
def priority(letter):
    if letter.isupper():
        return (ord(letter)-64)+26
    else:
        return (ord(letter)-96)


def solve_p1(x):
    total = 0
    for l, r in x:
        sub = list(set(l) & set(r))
        assert len(sub) == 1
        el  = sub[0]
        val = priority(el)
        print(el, val)
        total += val

    return total

solve_p1(parse_p1(test_input))
```

```python
def solve_p2(x):
    m = len(x)
    assert m % 3 == 0
    k = m // 3

    total = 0

    for idx in range(k):
        p1, p2 = idx * 3, idx * 3 + 3
        grp = x[p1:p2]

        s = None
        for l, r in grp:
            if s is None:
                s = set(l) | set(r)
            else:
                s = (set(l) | set(r)) & s
        assert len(s) == 1
        el  = list(s)[0]
        val = priority(el)
        print(el, val)
        total += val

    return total

solve_p2(parse_p1(test_input))
```
